Vamos us√°-lo para **gera√ß√£o de texto**, completando frases e respondendo perguntas simples.

```python
from transformers import pipeline, set_seed
```

## 2Ô∏è‚É£ Carregar o modelo DistilGPT-2

```python
# Carregar o pipeline de gera√ß√£o de texto
generator = pipeline("text-generation", model="distilgpt2")

# Definir semente para resultados reprodut√≠veis
set_seed(42)
```

## 3Ô∏è‚É£ Primeira gera√ß√£o de texto

```python
prompt = "Artificial intelligence is transforming the world because"
result = generator(prompt, max_length=50, num_return_sequences=1)

print(result[0]["generated_text"])
```

## 4Ô∏è‚É£ Gerando m√∫ltiplas varia√ß√µes

```python
prompt = "The future of deep learning is"
results = generator(prompt, max_length=60, num_return_sequences=3, do_sample=True, top_k=50)

for i, r in enumerate(results):
    print(f"\nüîπ Gera√ß√£o {i+1}:\n{r['generated_text']}")
```

## 5Ô∏è‚É£ Simulando Perguntas e Respostas simples

```python
questions = [
    "Q: What is deep learning?\nA:",
    "Q: Who created the theory of relativity?\nA:",
    "Q: What is the capital of Brazil?\nA:"
]

for q in questions:
    answer = generator(q, max_length=40, num_return_sequences=1)
    print(f"\n{answer[0]['generated_text']}")
```

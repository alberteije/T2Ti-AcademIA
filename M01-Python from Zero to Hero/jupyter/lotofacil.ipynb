## Carregando os Dados

```python
import pandas as pd

# Carregar os dados
caminho_arquivo = 'lotofacil-resultados.xlsx'
df_lotofacil = pd.read_excel(caminho_arquivo)

# Exibir as primeiras linhas do dataframe
df_lotofacil.head()
```

## Distribuição de Números Sorteados

> A frequência pode ajudar a determinar quais números são quentes ou frios, dando ao analista uma base para decidir quais números escolher. Mesmo que essa decisão não altere as probabilidades reais, ela pode influenciar estratégias de jogadores que acreditam nessas tendências.

<html>
<h3>.ravel():</h3>
<p>
Achata (flattens) a matriz em um vetor unidimensional.
<br>
Por exemplo, se os dados do DataFrame eram algo como:
</p>
<code>
[[1, 2, 3],
 [4, 5, 6]]
</code>
</p>
O resultado após ravel() seria:
</p>
<code>
[1, 2, 3, 4, 5, 6]
</code>
</html>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Transformando as colunas em uma única lista de números
todos_os_numeros = pd.Series(df_lotofacil.values.ravel())

# Contando a frequência de cada número
frequencia = todos_os_numeros.value_counts().sort_values(ascending=False)

# Gráfico de barras
plt.figure(figsize=(10, 6))
frequencia.plot(kind='bar', color='skyblue')
# sns.barplot(x=frequencia.index, y=frequencia.values, hue=frequencia.index, dodge=False, palette='viridis', legend=False)
plt.title('Frequência dos Números Sorteados')
plt.xlabel('Número')
plt.ylabel('Frequência')
plt.xticks(rotation=90)
plt.show()
```

---

## Análise de Padrões de Números - Soma

> Vamos explorar se existem padrões, como a soma dos números sorteados, para verificar se há alguma tendência. Se você está desenvolvendo estratégias para apostar, compreender a distribuição da soma pode ajudar a evitar combinações de números cujas somas estão muito fora da média, já que essas combinações podem ser menos prováveis. Apostar em números cuja soma esteja dentro da faixa mais comum pode aumentar as chances de o seu bilhete estar dentro de um padrão mais frequente.

```python
# Calcular a soma dos números sorteados
df_lotofacil['soma'] = df_lotofacil[[f'bola {i}' for i in range(1, 16)]].sum(axis=1)

# Criar um histograma da soma
plt.figure(figsize=(12, 6))
sns.histplot(df_lotofacil['soma'], bins=30, kde=True, color='blue')
plt.title('Distribuição da Soma dos Números Sorteados')
plt.xlabel('Soma dos Números')
plt.ylabel('Frequência')
plt.grid(axis='y')
plt.show()
```

---

## Gráfico de Calor para Combinações de Números
> O heatmap é uma representação visual em que a intensidade da cor em cada célula corresponde ao valor de uma combinação de elementos.
>
> **Números "Quentes"**: São os números que aparecem com maior frequência nos sorteios. Muitas pessoas acreditam que esses números têm mais chances de serem sorteados novamente, por estarem "em alta".
>
> **Números "Frios"**: São os números que aparecem menos frequentemente, ou que não são sorteados há muito tempo. Algumas pessoas escolhem esses números acreditando que, por terem sido sorteados menos vezes, estão "atrasados" e têm maior chance de aparecer em um próximo sorteio.

```python
plt.figure(figsize=(10, 6))
sns.heatmap(frequencia.values.reshape(1, -1), annot=False, cmap='coolwarm', cbar=True, xticklabels=frequencia.index)
plt.title('Distribuição de Calor dos Números Sorteados')
plt.xlabel('Número')
# plt.xticks(rotation=90) 
plt.show()
```

## Análise de Números Ímpares e Pares

> Agora, vamos analisar a proporção de números ímpares e pares que foram sorteados.
>
> Ao revisar sorteios anteriores, muitos jogadores percebem que combinações com extremos (como todos os números pares ou todos os números ímpares) são raras. Combinações que misturam números pares e ímpares aparecem mais frequentemente.

```python
# Contar ímpares e pares
números = df_lotofacil[[f'bola {i}' for i in range(1, 16)]]
impares = (números % 2 != 0).sum().sum()
pares = (números % 2 == 0).sum().sum() 

# Criar um gráfico de pizza
labels = ['Ímpares', 'Pares']
sizes = [impares, pares]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['orange', 'lightblue'])
plt.title('Proporção de Números Ímpares e Pares na Lotofácil')
plt.show()
```

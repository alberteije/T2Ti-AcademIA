## **2. Carregando Dados**

### Importando as Bibliotecas Necessárias

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Configurando o estilo dos gráficos com o Seaborn

```python
sns.set_theme(style="whitegrid")
```

### Leitura do arquivo do PROUNI

```python
file_path = '../pandas/ProuniRelatorioDadosAbertos2020.csv'
df = pd.read_csv(file_path, sep=';', encoding='latin1')
```

### Exibindo as primeiras linhas do DataFrame

```python
df.head()
```

## **3. Análise Exploratória**

### Exibindo informações gerais sobre o DataFrame

```python
df.info()
```

### Verificando valores ausentes

```python
df.isnull().sum()
```

### Exibindo as dimensões do DataFrame

```python
print(f"O DataFrame possui {df.shape[0]} linhas e {df.shape[1]} colunas.")
```

---

## **4. Manipulação de Dados**

### Renomeando as colunas para melhorar a legibilidade

```python
df.rename(columns={
  'SEXO_BENEFICIARIO': 'Sexo',
  'MODALIDADE_ENSINO_BOLSA': 'Modalidade',
  'TIPO_BOLSA': 'TipoBolsa'
}, inplace=True)

df.head()
```

### Substituindo valores ausentes na coluna NOME_CURSO_BOLSA

```python
df = df.fillna("AUSENTE")
```

### Verificando novamente os valores ausentes

```python
df.isnull().sum()
```

---

## **5. Estatísticas Descritivas**

### Distribuição de bolsas por sexo

```python
distribuicao_por_sexo = df['Sexo'].value_counts()
distribuicao_por_sexo_proporcao = df['Sexo'].value_counts(normalize=True) * 100
print("-------------------------------------------------------------------------")
print("Distribuição por Sexo")
print("-------------------------------------------------------------------------")
print(distribuicao_por_sexo)
print(distribuicao_por_sexo_proporcao)
```

### Distribuição das bolsas por Modalidade de Ensino

```python
modalidades = df['Modalidade'].value_counts()
print("\nDistribuição por Modalidade:")
print(modalidades)
```

---

## **6. Visualizações**

### Gráfico de barras: Distribuição por Sexo

```python
plt.bar(distribuicao_por_sexo.index, distribuicao_por_sexo.values, color=['skyblue', 'salmon'])
plt.title("Distribuição de Bolsas por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Quantidade")
plt.show()
```

### Gráfico de pizza: Distribuição por Modalidade

```python
plt.pie(modalidades.values, labels=modalidades.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title("Distribuição de Bolsas por Modalidade")
plt.show()
```

### Gráfico de Dispersão: Distribuição de Bolsas por Idade

```python
df['Idade'] = pd.to_datetime(df['DATA_NASCIMENTO'], errors='coerce').apply(lambda x: 2024 - x.year if pd.notnull(x) else None)
plt.scatter(df['Idade'], range(len(df)), alpha=0.1, color='purple')
plt.title("Distribuição de Bolsas por Idade")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()
```
